```python
# First let's eliminate the lines which are completely empty with the DROPNA function if they exist
import numpy as np
import pandas as pd
data=pd.read_csv('titanic-passengers.csv', sep=";")
data.dropna(axis=1, how='all', thresh=None,inplace= False)
data.isnull().sum().sum()
```

```python
#Replace empty age values with mode
import numpy as np
import pandas as pd
data['Age'].fillna(data['Age'].mode(), inplace=True)
data.head(10)
```

```python
#Identify the most used cabins
nb=len(data['Cabin'])
print(nb)
print(data["Cabin"].value_counts())
```

```python
# Let's replace the voids with the most used cabin
data['Cabin'].fillna('G6', inplace=True)
data.head(10)
```
