```python
#Split the dataset, Train test aned Linear regression for the feature: yr_built

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

x=data['yr_built'].values[:,np.newaxis]
y=data['price'].values
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.20, random_state=30)

model=LinearRegression()
model.fit(x_train,y_train)
predicted=model.predict(x_test)
print("MSE", mean_squared_error(y_test,predicted))
print("R squared",metrics.r2_score(y_test,predicted))
```

```python
# Plot the linear regression. 'yr_built'

plt.scatter(x,y,color='r')
plt.title('Linear Regression')
plt.ylabel('price')
plt.xlabel('yr_built')
plt.plot(x,model.predict(x), color='k')
plt.show
```

Exercice8 Apply Polynomial regression and compare it to the linear and multilinear regression. 

```python

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression



x=data[['bedrooms','bathrooms','sqft_living','grade','sqft_above','sqft_living15']]
y=data['price'].values
x_train, x_test,y_train, y_test = train_test_split(x, y,test_size=0.35, random_state= 30)

lg=LinearRegression()
poly=PolynomialFeatures(degree=2)


#Transformation de l'input Ã  utiliser pour l'entreianement du model
x_train_fit = poly.fit_transform(x_train)

# Entrainement du model
lg.fit(x_train_fit, y_train)

#Transformation de l'input Ã  utiliser pour la prdiction.
x_test_fit= poly.fit_transform(x_test)

# Prediction du model.
Predicted = lg.predict(x_test_fit)

# evaluation du model
print("MSE", mean_squared_error(y_test,Predicted))
print("R squared",metrics.r2_score(y_test,Predicted))
```

#Polynomial regression in degree 2 presents a better model than that linear and multilinear regression with an R squad
