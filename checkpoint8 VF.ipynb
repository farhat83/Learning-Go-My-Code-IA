### CHECKPOINT 8: Association Rules


```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]

print(dataset)
```


In order to use apriori function, we need to transform our dataset into a one-hot-encoded Dataframe.

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules 


te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
#Support Code
#Letâ€™s select itemsets with a minimum of 60% Support



#Support with column names
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
#Confidence Code
#In case we want to extract rules based on other metrics like confidence, we can use association_rules from mlxtend.frequent_patterns librar

from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%
```

```python
# Now let's see "lift"


association_rules(frequent_itemsets,metric="lift",min_threshold=1)
```


will list the rules that we have observed

```python
association_results = pd.DataFrame(association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7))
print(association_results)
```

```python
rows=association_results.shape[0]
columns=association_results.shape[1]

for x in range(rows):
    print("Rule: " + str(association_results.iloc[x,0]) + " -> " + str(association_results.iloc[x,1]) )
    print("Support: " + str(association_results.iloc[x,4]))
    print("Confidence: " + str(association_results.iloc[x,5]))
    print("Lift: " + str(association_results.iloc[x,6]))
    print("=====================================")
```
